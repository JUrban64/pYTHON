```python
szn1 = {}
with open("/Users/jachymurban/Downloads/3 hustota/staty_lidi.log", "r", encoding="utf-8") as file1: 
    for x in file1: 
        stat, lidi = x.strip().split(": ")
        szn1[stat] = lidi 

szn2 = {}
with open("/Users/jachymurban/Downloads/3 hustota/staty_plochy.log", "r", encoding="utf-8") as file2: 
    for x in file2: 
        stat, plocha = x.strip().split(": ")
        szn2[stat] = plocha 

data = [ ] 
from collections import namedtuple
informace = namedtuple("informace", ["stat", "lidi", "plocha", "hustota"]) ## vytvoření pojmenované n-tice, promena info, s nazven ibfrinace 


for stat in szn2: 
    plocha, lidi  = szn2[stat], szn1.get(stat, 0) ## syntax get = seznam.get(klíč, co dat když klič není)
    hustota = int(lidi) / float(plocha)
    data.append(informace(stat, lidi, plocha, hustota)) 



# seřadit podle husoty a printnout vysledek 
print("PODLE HUSTOTY")
for x in sorted(data, key=lambda y: y.hustota): 
    print(x.stat, x.hustota)

print("PODLE PLOCHY")
for y in sorted(data, key=lambda z: z.plocha): 
    print(y.stat, y.plocha)


```
