````python
import turtle
 
# otočení na sever
turtle.left(90)
 
# minimální vykreslovaná délka větví
MIN = 10
 
def strom(délka):
    # ukončovací podmínka rekurze (tzv. kotva)
    if délka < MIN: return
    # jinak kreslíme další iteraci
    turtle.forward(délka)
    turtle.left(60)
    délka_větví = délka * 0.7   # nová délka větví z tohoto místa
    strom(délka_větví)          # levá část větví
    turtle.right(120)
    strom(délka_větví)          # pravá část větví
    turtle.left(60)
    turtle.backward(délka)
 
# spuštění vykreslování stromu
strom(100)
 
# počkání na zavření okna uživatelem
turtle.done()
````

Třída `turtle` je knihovna v jazyce Python, která umožňuje jednoduché kreslení grafiky pomocí „želvy“ — kurzoru, který se pohybuje po obrazovce a kreslí čáry podle pokynů. Je to skvělý nástroj pro výuku základů programování i tvorbu jednoduchých obrázků nebo vizualizací.

### Jak začít:
Nejdříve je nutné importovat modul `turtle`:
```python
import turtle
```

Poté můžeme vytvořit objekt želvy:
```python
zelva = turtle.Turtle()
```

#### Základní metody:
Tady jsou nejdůležitější metody, které můžeš použít s objektem `Turtle`:

### Pohyb želvy
- **`forward(distance)`** nebo **`fd(distance)`**: Posune želvu dopředu o zadanou vzdálenost.
- **`backward(distance)`** nebo **`bk(distance)`** nebo **`back(distance)`**: Posune želvu dozadu o zadanou vzdálenost.
- **`right(angle)`** nebo **`rt(angle)`**: Otočí želvu doprava o zadaný úhel (ve stupních).
- **`left(angle)`** nebo **`lt(angle)`**: Otočí želvu doleva o zadaný úhel (ve stupních).
- **`goto(x, y)`** nebo **`setpos(x, y)`** nebo **`setposition(x, y)`**: Pohne želvu na souřadnice (x, y).
- **`setx(x)`**: Nastaví x-ovou souřadnici želvy.
- **`sety(y)`**: Nastaví y-ovou souřadnici želvy.
- **`home()`**: Vrátí želvu do výchozí pozice (0, 0) a natočí ji na výchozí směr (směr doprava).
- **`circle(radius, extent=None, steps=None)`**: Kreslí kruh s daným poloměrem. Parametr `extent` určuje část kruhu, kterou má želva nakreslit (např. 180 pro půlkruh), a `steps` určuje počet úseků, z nichž je kruh složen.

### Stav želvy
- **`penup()`** nebo **`pu()`** nebo **`up()`**: Zvedne pero, takže želva přestane kreslit čáry.
- **`pendown()`** nebo **`pd()`** nebo **`down()`**: Položí pero, takže želva začne opět kreslit.
- **`pensize(width)`** nebo **`width(width)`**: Nastaví tloušťku kreslící čáry.
- **`pencolor(color)`**: Nastaví barvu čáry (může být zadána jménem barvy jako string, např. `'red'`, nebo jako RGB hodnota).
- **`fillcolor(color)`**: Nastaví barvu výplně.
- **`begin_fill()`**: Začne vyplňování tvarů.
- **`end_fill()`**: Ukončí vyplňování tvarů.
- **`isdown()`**: Vrátí `True`, pokud je pero dole (kreslí).
- **`isup()`**: Vrátí `True`, pokud je pero nahoře (nekreslí).

### Ovládání vzhledu želvy
- **`shape(name)`**: Změní tvar kurzoru želvy (může být `'turtle'`, `'arrow'`, `'circle'`, `'square'`, `'triangle'`, `'classic'`).
- **`shapesize(stretch_wid=None, stretch_len=None, outline=None)`**: Nastaví velikost tvaru želvy.
- **`speed(speed)`**: Nastaví rychlost pohybu želvy. Hodnoty jsou mezi 1 (pomalá) a 10 (velmi rychlá), nebo `'fastest'`, `'fast'`, `'normal'`, `'slow'`, `'slowest'`.

### Stav želvy (informace o pozici)
- **`position()`** nebo **`pos()`**: Vrátí aktuální souřadnice želvy jako tuple (x, y).
- **`heading()`**: Vrátí aktuální úhel natočení želvy (ve stupních).
- **`towards(x, y)`**: Vrátí úhel směru k bodu (x, y).
- **`xcor()`**: Vrátí aktuální x-ovou souřadnici želvy.
- **`ycor()`**: Vrátí aktuální y-ovou souřadnici želvy.

### Řízení obrazovky
- **`turtle.Screen()`**: Vytvoří objekt obrazovky pro vykreslování grafiky.
- **`turtle.bgcolor(color)`**: Nastaví barvu pozadí obrazovky.
- **`turtle.clearscreen()`** nebo **`turtle.clear()`**: Vymaže všechny kresby na obrazovce.
- **`turtle.hideturtle()` nebo **`ht()`**: Skryje kurzor želvy.
- **`turtle.showturtle()` nebo **`st()`**: Znovu zobrazí kurzor želvy.
- **`turtle.reset()`**: Vymaže kresbu a nastaví želvu na výchozí pozici.
- **`turtle.bye()`**: Uzavře okno s kresbou.

### Ukázkový program:
```python
import turtle

zelva = turtle.Turtle()

# Nastavení barvy čáry a výplně
zelva.pencolor("blue")
zelva.fillcolor("yellow")

# Začátek vyplňování
zelva.begin_fill()

# Kreslení čtverce
for _ in range(4):
    zelva.forward(100)
    zelva.right(90)

# Konec vyplňování
zelva.end_fill()

turtle.done()
```

Tento kód nakreslí žlutě vyplněný modrý čtverec.
